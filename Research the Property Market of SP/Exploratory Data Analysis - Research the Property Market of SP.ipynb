```python
def area_category(row):
    total_area = row['total_area']
    if total_area < 31:
        return "<30"
    if total_area < 61:
        return "<60"
    if total_area < 101:
        return "<100"
    if total_area < 201:
        return "<200"
    if total_area < 401:
        return "<400"
    else:
        return "<900"

data['total_area_category'] = data.apply(area_category, axis=1)
data.head()
```

```python
data['living_area'] = (data['living_area'].
fillna(data.groupby('total_area_category')['living_area'].transform('median')))

data['kitchen_area'] = (data['kitchen_area'].
fillna(data.groupby('total_area_category')['kitchen_area'].transform('median')))
```

### Группировка по этажам

<div style="border:solid blue 1px; padding: 20px"> 
Сгруппируем данные по этажам: первый, последний, другой. 

```python
def floor_category(row):
    floor = row['floor']
    floors_total = row ['floors_total']
    if floor == floors_total:
        return "последний"
    if floor == 1:
        return "первый"
    else:
        return "другой"

data['floor_category'] = data.apply(floor_category, axis=1)
data.head()
```

### Расчет соотношений жилой и площади кухни к общей площади

<div style="border:solid blue 1px; padding: 20px">

```python
data['living_total_area'] = round(data['living_area'] / data['total_area'],2)
data['kitchen_total_area'] = round(data['kitchen_area'] / data['total_area'],2)
```

Самые высокие цены на недвижимость были в 2019 году в центре и в 2014 году в остальных районах города.
